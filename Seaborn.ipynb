House price prediction


```python
from sklearn.datasets import fetch_california_housing
california=fetch_california_housing()
house=pd.DataFrame(california.data,columns=california.feature_names)
house['PRICE']=california.target
house.head()
```

 Distribution of data

```python
sns.displot(house['PRICE'])
```

```python
sns.distplot(house['PRICE'])
```

Co-relation

1.Positive Corelation

2.Negative Corelation

```python
#Construct a heat map
corelation=house.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corelation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
```
